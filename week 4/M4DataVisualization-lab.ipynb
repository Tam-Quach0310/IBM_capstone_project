## Demo: How to work with database


Download database file.


```python
import urllib.request
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite'
filename = 'm4_survey_data.sqlite'
urllib.request.urlretrieve(url, filename)
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.head(1)
```

```python
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df, x='ConvertedComp', bins=40)
plt.title('Distribution of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.figure(figsize=(6,4))
sns.boxplot(df, y='Age', orient='v')
plt.title('Distribution of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.regplot(x='Age', y='WorkWeekHrs', data=df)
plt.title('Relationship between Age and Working hours per Week')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(6,6))
sns.scatterplot(df, x='WorkWeekHrs', y='CodeRevHrs', size='Age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT * FROM DatabaseDesireNextYear
"""

df_database = pd.read_sql_query(QUERY,conn)
df_top5 = df_database['DatabaseDesireNextYear'].value_counts().head()
plt.pie(df_top5, labels=df_top5.index, autopct='%1.1f%%')
plt.title('Top 5 databases that respondents wish to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_agegroup = df[((df['Age']>=30) & (df['Age']<=35))]
med_work = df_agegroup['WorkWeekHrs'].median()
med_code = df_agegroup['CodeRevHrs'].median()
df_med = pd.DataFrame({'med work': med_work, 'med code': med_code}, index=[0])

df_med.plot.bar(stacked=True, figsize=(4,4))
plt.legend(['median WorkWeekHrs', 'median CodeRevHrs'], bbox_to_anchor=(1,1))
plt.ylabel('median')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_45to60 = df[((df['Age']>=45) & (df['Age']<=60))]
df_med45to60 = df_45to60.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10,6))
sns.lineplot(x=df_med45to60.index, y=df_med45to60.values)
plt.title('Median ConvertedComp for all ages from 45 to 60')
plt.ylabel('Median')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
plt.figure(figsize=(6,3))
sns.countplot(y=df['MainBranch'], orient='h')
plt.show()
```
